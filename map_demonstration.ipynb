```python
import folium
```

```python
m = folium.Map(location=[60.355, 5.21], zoom_start=12)
 
pickups = [[60.382878, 5.055929], [60.352223, 5.087813], [60.388651, 4.993453], [60.261564, 4.982022], [60.356561, 5.121333], [60.368083, 5.153025], 
[60.354085,5.106085], [60.390343,5.039608],[60.375988, 5.106568], [60.354616, 5.152274]]

deliveries = [[60.395409, 5.326163], [60.379099, 5.264793], [60.290962, 5.258676], [60.363501, 5.352355]]

for i in pickups:
    folium.CircleMarker(location = i, radius = 4, color = 'blue', fill_color ='blue').add_to(m)

for i in deliveries:
    folium.CircleMarker(location = i , radius = 4, color = 'red', fill_color ='red').add_to(m)



m
```
